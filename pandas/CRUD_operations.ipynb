```python
import pandas as pd

df = pd.read_csv("orders.csv")
```

```python
#.loc means location and it is used to access a group of rows and columns by labels or a boolean array.
#labels means the names of the rows and columns. we can change the corresponding labels to something else too

```

```python

df.loc[df["CustomerName"] == "Anna Ivanova"]
```

```python
#change the qty ordered by ana to 10


df.loc[df["CustomerName"] == "Anna Ivanova", "Quantity"] = 12
df[df["CustomerName"] == "Anna Ivanova"]
```

```python
#another example
#change the country name from USA to United States

df.loc[df["Country"]=="USA", "Country"] = "United States"
df[df["Country"]=="United States"]
```

```python
df["Country"]=df["Country"].str.upper() #changing all country names to upper case

df[df["Country"]=="CANADA"] #filtering based on a condition
```

```python
#id you want to delete we use drop
df.head()
```

```python
#to delete an entire row of data we use the index of that row
df = df.drop(2) 
```

```python
df.dropna()

# THIS WILL DELETE ALL THE ROWS WHERE THERE IS AT LEAST ONE NULL VALUE 


```

```python

df.fillna({"OrderID": 0}, inplace=True) #this will return a new dataframe with the null values filled with 0
```

```python
#renaming columns
df.rename(columns={"OrderID": "Order_ID"}, inplace=True)
```

```python
#saving a file

df.to_csv("modified_orders.csv", index=False)
```
